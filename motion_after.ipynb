```python
import pandas as pd
import ast

# 1) Load the data
df = pd.read_csv('/content/full_output.csv')

# 2) Columns to inspect
cols = [
    'rt1','rt2',
    'direction_of_motion1','direction_of_motion2',
    'direction_followup1','direction_followup2',
    'blink1','blink2'
]

# 3) Helper to parse the one-element lists into strings (or keep scalars as-is)
def unpack_cell(x):
    try:
        val = ast.literal_eval(x)
        # if it’s a list with one element, return that element
        if isinstance(val, list) and len(val)==1:
            return val[0]
        # if it’s a longer list, return the whole list
        return val
    except:
        # not a literal list—just return as-is
        return x

# Apply unpacking
for c in cols:
    df[c + '_unpacked'] = df[c].fillna('').astype(str).apply(unpack_cell)

# 4) Gather uniques
unique_responses = {}
for c in cols:
    series = df[c + '_unpacked']
    # if the unpacked is sometimes a list, flatten it
    all_vals = []
    for v in series:
        if isinstance(v, list):
            all_vals.extend(v)
        else:
            all_vals.append(v)
    # drop blanks and sort
    uniques = sorted([u for u in set(all_vals) if str(u).strip()!=''])
    unique_responses[c] = uniques

# 5) Print them out
#for c, vals in unique_responses.items():
    #print(f"\nUnique values in {c} ({len(vals)}):")
    #for v in vals:
        #print(" -", v)
```

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# 1) Load
df = pd.read_csv('/content/full_output.csv')

# 2) Define your known categories for each column
categories = {
    'direction_of_motion1': [
        'The dots did not seem to move at all',
        'The dots seemed to move but I could not sense a direction of movement',
        'The dots seemed to move in another way',
        'The dots seemed to move towards the centre of the screen',
        'The dots seemed to move towards the edges of the screen',
    ],
    'direction_of_motion2': [
        'The dots did not seem to move at all',
        'The dots seemed to move but I could not sense a direction of movement',
        'The dots seemed to move in another way',
        'The dots seemed to move towards the centre of the screen',
        'The dots seemed to move towards the edges of the screen',
    ],
    'direction_followup1': [
        'I experienced the dots as moving and changing position',
        'I experienced the dots as moving, but not changing position',
    ],
    'direction_followup2': [
        # intentionally empty → everyone will be No response
    ],
    'blink1': [
        'I blinked more than once',
        'I blinked once',
        'I did not blink',
    ],
    'blink2': [
        'I blinked more than once',
        'I blinked once',
        'I did not blink',
    ],
}

n = len(df)

for col, cats in categories.items():
    title = col.replace('_',' ').title()

    # 3) Prepare responses, filling missing or unexpected with "No response"
    resp = df[col].fillna('No response').astype(str)
    resp = resp.where(resp.isin(cats), 'No response')

    # 4) Create a Categorical with your order + "No response" last
    ordered = cats + ['No response']
    resp_cat = pd.Categorical(resp, categories=ordered, ordered=True)

    # 5) Count and percent
    counts = pd.value_counts(resp_cat, sort=False)
    pct    = counts / n * 100

    # 6) Plot side-by-side bar charts
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), sharey=False)

    # Raw counts
    bars1 = ax1.bar(counts.index, counts.values, color='C0')
    ax1.set_title(f'{title} — Counts')
    ax1.set_ylabel('Number of participants')
    ax1.set_xticklabels(counts.index, rotation=45, ha='right', fontsize=8)
    for bar, val in zip(bars1, counts.values):
        ax1.text(bar.get_x() + bar.get_width()/2, val + 1,
                 f'{int(val)}', ha='center', va='bottom', fontsize=8)

    # Percentages
    bars2 = ax2.bar(pct.index, pct.values, color='C1')
    ax2.set_title(f'{title} — % of participants')
    ax2.set_ylim(0, 100)
    ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
    ax2.set_xticklabels(pct.index, rotation=45, ha='right', fontsize=8)
    for bar, val in zip(bars2, pct.values):
        ax2.text(bar.get_x() + bar.get_width()/2, val + 1,
                 f'{val:.1f}%', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()

    # 7) Optional pie chart of percentages
    fig, ax = plt.subplots(figsize=(6, 6))
    wedges, texts, autotexts = ax.pie(
        pct.values, labels=None,
        autopct=lambda pct: f'{pct:.1f}%' if pct > 0 else '',
        startangle=90
    )
    ax.set_title(f'{title} — % of participants')
    ax.legend(wedges, pct.index, title=title,
              bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    plt.tight_layout()
    plt.show()
```
